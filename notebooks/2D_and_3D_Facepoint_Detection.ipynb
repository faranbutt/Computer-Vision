```python
import face_alignment
from skimage import io
import cv2
```

```python
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType.TWO_D,flip_input=True,device='cpu')
```

```python
inp = cv2.imread('/content/close photo.jpg')
```

```python
preds = fa.get_landmarks(inp)[0]
```

```python
preds.shape
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

fig,ax = plt.subplots(figsize=(5,5))
plt.imshow(cv2.cvtColor(inp, cv2.COLOR_BGR2RGB))
ax.scatter(preds[:,0],preds[:,1],marker = '+',c='r')
plt.show()
```
